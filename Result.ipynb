```python
def getDprime(df,mp):
    percision = str(mp)+'_Percision'
    recall = str(mp) + '_Recall'
    df[str(mp)+'_dprime'] = -(1-df[percision]).apply(norm.ppf)+df[recall].apply(norm.ppf)
    return df
for i in [1,5,30,50,90]:
    getDprime(finedbrand,i)
#finedbrand
```

```python
import re
def process_df(indf,name,sr):
    df = pd.melt(indf, id_vars=['UserID'], value_vars=['1_Accuracy', '5_Accuracy','30_Accuracy'],
            var_name='MP', value_name='Accruacy')
    percision = pd.melt(indf, id_vars=['UserID'], value_vars=['1_Percision', '5_Percision','30_Percision'],
            var_name='MP', value_name='Precision')
    recall = pd.melt(indf, id_vars=['UserID'], value_vars=['1_Recall', '5_Recall','30_Recall'],
            var_name='MP', value_name='Recall')
    df['MP'] = df['MP'].apply(lambda x: int(re.findall(r'\d+',x)[0]))
    df['Precision'] = percision['Precision']
    df['Recall'] = recall['Recall']
    df['Approach'] = [name for x in range(len(df))]
    df['SplitRatio'] = sr
    return df
```

```python
processeddfs = [process_df(randdfs[key],key) for key,value in randdfs.items()]
resultscombine = pd.concat(processeddfs)
resultscombine.to_csv('combined_result.csv',index = False)
```
